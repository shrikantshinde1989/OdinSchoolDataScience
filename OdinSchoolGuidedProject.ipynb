```python
#Encoding (label encoding on the business travel) 

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Label Encoding'] = le.fit_transform(df['BusinessTravel'])
#df['Label Encoding'].value_counts()
#df.shape
df.head()
df[['Label Encoding' ,'BusinessTravel']]
```

```python
#Want to select all the numerical columns and scale the data along with the business travel column(encoded).

#df.info()

# to select only numerical columns or feature

X = df.select_dtypes(include='number')
#X.info()

# Another way
#num_cols = df.select_dtypes(include='number').columns
#X = df.loc[:,num_cols]
#X

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
#Apply PCA on the numerical column selected above
from sklearn.decomposition import PCA
pca = PCA(n_components=27)
fit = pca.fit(X)
```

```python
plt.figure(figsize=(10,5))
PC_values = np.arange(pca.n_components_) + 1
#print(PC_values) it start from 0 that's why we add here + 1
plt.plot(PC_values,pca.explained_variance_ratio_,'o-',linewidth = 2,color = 'blue')
plt.xticks(range(1,28))
plt.title("Scree Plot with Elbow Plot",fontsize=12)
plt.xlabel("Principal Component")
plt.ylabel("Variance Explained")
```

```python
pca.explained_variance_ratio_
```

### Bar plot for PCA

```python
per_var = np.around(pca.explained_variance_ratio_*100,decimals=1)
#print(per_var)
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
#labels
plt.figure(figsize=(15,5))
plt.bar(x=range(1,len(per_var)+1),height=per_var,tick_label=labels)
plt.title("SCree Plot by Bar Plot",fontsize=15)
plt.xlabel("Principal Component")
plt.ylabel("Percentage of Explained Variance")
```
