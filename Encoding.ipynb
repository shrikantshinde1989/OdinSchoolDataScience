### Encoding
- Types of Encoding
    - Dummy Encoding
    - One Hot Encoding
    - Ordinal Encoding(Rank Encoding)
    - Label Encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df = pd.DataFrame(data=["India","India","Pakistan","Bangladesh","Srilanka","Thailand","Afganistan","Myanmar"],columns=["Country"])
df
```

```python
le = LabelEncoder()
df["Label Encoding"] = le.fit_transform(df["Country"])
df
```

```python
levels = {"India":1,"Thailand":2,"Bangladesh":3,"Srilanka":4,"Myanmar":5,"Pakistan":6,"Afganistan":7}
levels
```

```python
df["Ordinal Encoding"] = df["Country"].map(levels)
df
```

```python
#Dummy Encoding
pd.get_dummies(df,columns=["Country"],prefix='sp')
```
